<h3> Q2: Select a number randomly with probability proportional to its magnitude from the given array of n elements</h3>

consider an experiment, selecting an element from the list A randomly with probability proportional to its magnitude.
assume we are doing the same experiment for 100 times with replacement, in each experiment you will print a number that is selected randomly from A.

<pre>
Ex 1: A = [0 5 27 6 13 28 100 45 10 79]
let f(x) denote the number of times x getting selected in 100 experiments.
f(100) > f(79) > f(45) > f(28) > f(27) > f(13) > f(10) > f(6) > f(5) > f(0)
</pre>

```python
import random
import pandas as pd

def input_fun(length) :
    
    num_list = []
    for l in range(length):
        num = int(input("enter "+ str(l+1) +" number :"))
        num_list.append(num)
    return num_list

def pick_a_number_from_list(A,length):
    sum = 0
    for i in range (length):
        sum = sum+A[i]
    new_A = []
    for i in range(length):
        result = A[i] / sum
        new_A.append(result)
    series = pd.Series(new_A)                               #reference taken fromm stack overflow
    replaced_value = series.cumsum(skipna = False)
    
    probability = random.uniform(0.0, 1.0)                        # reference taken from pynative.com  
    
    for l in range(length):
        if probability < replaced_value[l] :
            return l
        elif probability >replaced_value[1] and probability < replaced_value[l+1]:
            return l+1
                                                    
def sampling_based_on_magnitued():
    length  = int(input(" How many Number you want to enter ??"))
    num_list = input_fun(length)
    #A = [0,5,27,6,13,28,100,45,10,79]   
    #length_of_A = len(A)
    for i in range(1,100):
        number = pick_a_number_from_list(num_list,length)
        print(num_list[number])
        
sampling_based_on_magnitued() 
```

<h3> Q4: Students marks dashboard</h3>

Consider the marks list of class students given in two lists <br>
Students = ['student1','student2','student3','student4','student5','student6','student7','student8','student9','student10'] <br>
Marks = [45, 78, 12, 14, 48, 43, 45, 98, 35, 80] <br>
from the above two lists the Student[0] got Marks[0],  Student[1] got Marks[1] and so on. <br><br>

Your task is to print the name of students

<strong>a. Who got top 5 ranks, in the descending order of marks</strong> <br>
<strong>b. Who got least 5 ranks, in the increasing order of marks</strong><br>
<strong>d. Who got marks between  &gt;25th percentile &lt;75th percentile, in the increasing order of marks.</strong>

<pre>
Ex 1: 
Students=['student1','student2','student3','student4','student5','student6','student7','student8','student9','student10'] 
Marks = [45, 78, 12, 14, 48, 43, 47, 98, 35, 80]

a.

b.
student3 12
student4 14
student9 35
student6 43
student1 45

c.
student9 35
student6 43
student1 45
student7 47
student5 48

```python
from collections import Counter
import numpy as np

def input_data():
    Students = []
    Marks = []
    n = int(input("Enter the total number of Students : "))
    for i in range(0,n):
        i = str(input("Enter the name of student " +str(i+1)+" : "))
        Students.append(i)

    for j in range (0,n):
        j = int(input("Enter the Mark of student  " +str(j+1)+" :"))
        Marks.append(j)
   
    per_25 = np.percentile(Marks, 25)
    per_75 = np.percentile(Marks, 75)
    
    return Students,Marks,per_25,per_75

def Convert_to_dict (Students,Marks):
    new_students_marks = {}                       # converting two list into one dictionary 
    for key in Students:                          # reference taken from geekfromgeeks
        for value in Marks: 
            new_students_marks[key] = value 
            Marks.remove(value) 
            break
    return new_students_marks

def Top_Rank (new_students_marks):
    desc_sorted_dict = sorted(new_students_marks .items(), key=lambda x: x[1] , reverse=True)  
    k1= Counter(desc_sorted_dict)
    top_rank = k1.most_common(5) 
    print("Name of Students who got TOP 5 ranks :")
    for key,value in top_rank: 
        print(key)

def Least_Rank(new_students_marks):
    asc_sorted_dict = sorted(new_students_marks .items(), key=lambda x: x[1])
    #print(asc_sorted_dict)
    k2 = Counter(asc_sorted_dict)
    least_rank = k2.most_common(5) 

    print("\n Name of Students who got LEAST 5 ranks :")
    for key,value in least_rank: 
        print(key) 
    return asc_sorted_dict
        
def Percentile_Rank(asc_sorted_dict,per_25,per_75):    
    print("\n The students Who got marks between >25th percentile :"+str(per_25) +"<75th percentile :"+ str(per_75))
    print(*(" {}, {}\n ".format(students,marks) for (students, marks) in asc_sorted_dict if (marks >per_25 and marks < per_75)))
       
# functions calling     
Students,Marks,per_25,per_75 = input_data()
new_stud_marks = Convert_to_dict (Students,Marks)
#print(new_stud_marks)
Top_Rank (new_stud_marks)
asc_sorted_dict = Least_Rank(new_stud_marks)
Percentile_Rank(asc_sorted_dict,per_25,per_75)      
```

<h3> Q5: Find the closest points</h3>

Consider you are given n data points in the form of list of tuples like S=[(x1,y1),(x2,y2),(x3,y3),(x4,y4),(x5,y5),..,(xn,yn)] and a point P=(p,q) <br> your task is to find 5 closest points(based on cosine distance) in S from P

<br>Cosine distance between two points (x,y) and (p,q) is defined as $cos^{-1}(\frac{(x\cdot p+y\cdot q)}{\sqrt(x^2+y^2)\cdot\sqrt(p^2+q^2)})$
<pre>
Ex:

S= [(1,2),(3,4),(-1,1),(6,-7),(0, 6),(-5,-8),(-1,-1)(6,0),(1,-1)]
P= (3,-4)
<img src='https://i.imgur.com/vIFPOcG.jpg', width=300>
Output:
(6,-7)
(1,-1)
(6,0)
(-5,-8)
(-1,-1)
</pre>

```python
import math

S= [(1,2),(3,4),(-1,1),(6,-7),(0, 6),(-5,-8),(-1,-1),(6,0),(1,-1)]
P= (3,-4)

print("Cosine Distance between the two points :")
new_list = sorted(S, key= lambda x: math.acos((float(x[0]* P[0] + x[1] * P[1])) \
            / ((float(math.sqrt (x[0]* x[0] +x[1] * x[1])) * (math.sqrt (P[0]* P[0] +P[1] * P[1]))))) )  #geeksfromgeeks

print(new_list)
print("The closest points are :")
#new_list.sort()
print(new_list[:5])
```

</pre>
for a given string with comma seprate values, which will have both missing values numbers like ex: "_, _, x, _, _, _"

Ex: 
<pre>

Input2: "40,_,_,_,60"

```python
def curve_smoothing (input_string):
    #string_split = []
    string_split = input_string.split(",")
    ser_number = [i for i, value in enumerate(string_split) if value != '_']           # reference taken from stack overflow
    
    if ser_number[0] != 0:
        ser_number = [-1] + ser_number
    
    if ser_number[-1] != len(string_split)-1:
        ser_number = ser_number + [-1]    
    
    for (start, end) in zip(ser_number[:-1], ser_number[1:]): 
        if start == -1:
            value = float(string_split[end])/(end+1)
            for i in range(start+1,end+1):
                string_split[i] = int( value) 
        elif end == -1:
            value = float(string_split[start])/(len(string_split)-start)
            for i in range(start, len(string_split)):
                string_split[i] = int( value ) 
        else:
            value = (int(string_split[start])+int(string_split[end]))/(end-start+1)
            for i in range(start,end+1):
                string_split[i] = int( value )         
    return string_split
    
input_string_1 = "_,_,_,24 "
input_string_2 = "40,_,_,_,60"
input_string_3 = "80,_,_,_,_"
input_string_4 = "_,_,30,_,_,_,50,_,_"

split1 = curve_smoothing (input_string_1)
print(input_string_1 + " :  " +str(split1))
split2 = curve_smoothing (input_string_2)
print(input_string_2 + " :  " +str(split2))
split3 = curve_smoothing (input_string_3)
print(input_string_3 + " :  " +str(split3))
split4 = curve_smoothing (input_string_4)
print(input_string_4 + " :  " +str(split4))
```

<h3> Q9: Operations on sentences </h3>

You will be given two sentances S1, S2 your task is to find 
<pre>
a. Number of common words between S1, S2
b. Words in S1 but not in S2
c. Words in S2 but not in S1
</pre>

Ex: 
<pre>
S1= "the first column F will contain only 5 unique values"
S2= "the second column S will contain only 3 unique values"
Output:
a. 7
b. ['first','F','5']
c. ['second','S','3']
</pre>

```python
def string_features(str1, str2):
    split_1 = str1.split(" ")
    #print(split_1)
    split_2 =  str2.split(" ")
    #print(split_2)
    count = 0
    for s1 in range(len(split_1)):
        for s2 in range(len(split_2)):
            if split_1[s1] == split_2[s2]:
                count +=1
                #print(split_1[s1])          
    return count,split_1,split_2


str1 = input("Enter the first string")
str2 = input("Enter the second string ")
count,split_1,split_2 = string_features (str1,str2)
print("Number of common words between Str1, Str2 :",count)

set_1_uncom = [ x for x in split_1 if x not in split_2 ]
print(set_1_uncom)

set_2_uncom = [ x for x in split_2 if x not in split_1 ]
print(set_2_uncom)
```

Your task is to find the value of $f(Y,Y_{score}) = -1*\frac{1}{n}\Sigma_{for each Y,Y_{score} pair}(Ylog10(Y_{score})+(1-Y)log10(1-Y_{score}))$ here n is the number of rows in the matrix
<pre>
Ex:
[[1, 0.4], [0, 0.5], [0, 0.9], [0, 0.3], [0, 0.6], [1, 0.1], [1, 0.9], [1, 0.8]]
output:
0.44982
</pre>
$\frac{-1}{8}\cdot((1\cdot log_{10}(0.4)+0\cdot log_{10}(0.6))+(0\cdot log_{10}(0.5)+1\cdot log_{10}(0.5)) + ... + (1\cdot log_{10}(0.8)+0\cdot log_{10}(0.2)) )$

```python
import math
matrix_list = [[1, 0.4], [0, 0.5], [0, 0.9], [0, 0.3], [0, 0.6], [1, 0.1], [1, 0.9], [1, 0.8]]

def compute_log_loss(matrix_list):
    loss = 0
    for m in range(len(matrix_list)):
        loss = loss + ((matrix_list[m][0] * math.log (matrix_list[m][1],10)) \
                       + ((1- matrix_list[m][0]) * math.log((1 - matrix_list[m][1]),10)))  #formula math.log(a,base) from geeks
        #print(loss)
    loss1 = loss * (-1 / len(matrix_list)) 
    return loss1

loss = compute_log_loss(matrix_list)
print("loss :",loss)
```
